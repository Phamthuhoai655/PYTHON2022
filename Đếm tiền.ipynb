```python
count = 0
for i in range(1,50):
    for j in range(1,100-2*i):
        if (100 - (2*i + 3*j)) % 5 == 0 and ((100 - (2*i + 3*j))/5)>0:
            count += 1
            print("Cach {0}: 2*{1} + 3*{2} + 5*{3} = 100".format(count,i,j,(100 - (2*i + 3*j))//5))
        else:
            count += 0
        
print('Co ',count,' cach')
    
```
