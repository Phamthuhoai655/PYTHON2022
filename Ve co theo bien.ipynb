```python
import turtle

h=turtle.Turtle()
h.pensize(1)
h.speed(5)

height = 8
width = 4
h.pencolor('blue')
h.fillcolor('blue')
h.penup()
h.backward(height*30)
h.left(90)
h.forward(height*10)
h.right(90)
h.pendown()
h.begin_fill()
h.forward(height*10)
h.right(90)
h.forward(width*16)
h.right(90)
h.forward(height*10)
h.right(90)
h.forward(width*16)
h.end_fill()

h.pencolor('yellow')
h.right(90)
h.penup()
h.forward(height*10)
h.pendown()
h.fillcolor('yellow')
h.begin_fill()
h.forward(height*4)
h.right(90)
h.forward(width*16)
h.left(90)
h.forward(height*18)
h.right(90)
h.forward(width*8)
h.right(90)
h.forward(height*18)
h.left(90)
h.forward(width*16)
h.right(90)
h.forward(height*4)
h.right(90)
h.forward(width*16)
h.left(90)
h.forward(height*10)
h.right(90)
h.forward(width*8)
h.right(90)
h.forward(height*10)
h.left(90)
h.forward(width*16)
h.end_fill()

h.pencolor('blue')
h.penup()
h.right(90)
h.forward(width*8)
h.pendown()
h.fillcolor('blue')
h.begin_fill()
h.forward(height*18)
h.right(90)
h.forward(width*16)
h.right(90)
h.forward(height*18)
h.right(90)
h.forward(width*16)
h.penup()
h.backward(width*24)
h.right(90)
h.pendown()
h.forward(height*18)
h.right(90)
h.forward(width*16)
h.right(90)
h.forward(height*18)
h.right(90)
h.forward(width*16)
h.end_fill()

h.pencolor('blue')
h.left(90)
h.penup()
h.forward(width*8)
h.pendown()
h.fillcolor('blue')
h.begin_fill()
h.forward(height*10)
h.left(90)
h.forward(width*16)
h.left(90)
h.forward(height*10)
h.left(90)
h.forward(width*16)
h.end_fill()

turtle.done()
```
